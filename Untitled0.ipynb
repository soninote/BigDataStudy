```python
#구글 코랩과 구글 드라이브 연동
from google.colab import drive
drive.mount('/gdrive', force_remount = True)
#필요한 라이브러리 가져오기
import requests as rq
import pandas as pd
#Geogoder API 연결하기
url_front = 'http://api.vworld.kr/req/address?'
url_params = 'service=address&request=getcoord&version=2.0&crs=epsg:4326&refine=true&simple=false&format=json&type=road' #주소는 따로 받고 형식을 xml -> json 으로 변경해줌
url_address = '&address='
url_key = '&key='
auth_key = '6D80D0CB-D1FE-3DD1-9C2F-6F8DD380CBCA'
#주소가 입력된 엑셀 파일 읽기
df = pd.read_excel('/gdrive/MyDrive/andong.xlsx')#df = pd.read_csv('')csv 파일일때
df
#위도와 경도값을 저장할 리스트 선언하기
latitude = []
logitude = []
#반복문을 이용하여 차례대로 주소를 읽어서 위도와 경도 알아내기
for i in range(len(df)): # 오류 range 제외시
  address = df['주소'][i]
  url = url_front+url_params+url_address+address+url_key+auth_key
  result = rq.get(url)
  json_data = result.json()

  if json_data['response']['status'] == 'OK':
    x = json_data['response']['result']['point']['x']
    y = json_data['response']['result']['point']['y']
    there = df['장소명'][i]
    print(f'{there} : {x}, {y}')
    logitude.append(x)
    latitude.append(y)
  else:
    df.drop(i, inplace = True) # 안나온행 삭제
    
#위도와 경도를 데이터 프레임에 컬럼으로 추가하기
df['위도'] = latitude
df['경도'] = logitude
# 장소명 컬럼을 인덱스로 설정
df.set_index('장소명', inplace = True)
#수정된 데이터 프레임을 엑셀 파일로 저장하기
df.to_excel('/gdrive/MyDrive/andong_1.xlsx')
```

내가 선정한 장소에 마커 표시하기

```python
#라이브러리 가져오기
import pandas as pd
import folium as fo

lat, log = 36.543682613, 128.800027098
zoom = 13
map = fo.Map(
    location= [lat, log],
    zoom_start= zoom
)
df = pd.read_excel('/gdrive/MyDrive/andong_1.xlsx')
for i in range(len(df)):
  thereLoc = df['위도'][i], df['경도'][i]
  fo.CircleMarker(
      thereLoc,
      radius = 20,
      color = '#8000FF',
      fill_color = '#8000FF'
  ).add_to(map)

  fo.Marker(
      thereLoc,
      tooltip = df['장소명'][i],
      icon = fo.Icon(color = 'red', icon = 'heart' )
  ).add_to(map)
# html 파일로 저장
map.save('/gdrive/MyDrive/andong.html')
```

folium 에서 제공하는 배경지 이용하기

```python
#라이브러리 가져오기
import pandas as pd
import folium as fo

lat, log = 36.543682613, 128.800027098
zoom = 13
map = fo.Map(
    location= [lat, log],
    zoom_start= zoom
)

#OpenStreeMap(기본), Staman Terrain, Stamen Tonner, Stamen Watercolor
#CartoDB positron, CartoDB dark_matter

# 배경지도 타일 설정하기
tiles = 'OpenStreetMap'
fo.TileLayer(tiles = tiles).add_to(map)

df = pd.read_excel('/gdrive/MyDrive/andong_1.xlsx')
for i in range(len(df)):
  thereLoc = df['위도'][i], df['경도'][i]
  fo.CircleMarker(
      thereLoc,
      radius = 20,
      color = '#8000FF',
      fill_color = '#8000FF'
  ).add_to(map)

  fo.Marker(
      thereLoc,
      tooltip = df['장소명'][i],
      icon = fo.Icon(color = 'red', icon = 'heart' )
  ).add_to(map)
map
```

구글 지도를 배경으로 설정하기



```python
#라이브러리 가져오기
import pandas as pd
import folium as fo

lat, log = 36.543682613, 128.800027098
zoom = 13
map = fo.Map(
    location= [lat, log],
    zoom_start= zoom
)


# 구글지도 배경 설정하기
googleMap = {
    'Google Maps' : fo.TileLayer(
        tiles = 'http://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'hello',
        overlay = True,
        control = True)
}
googleMap['Google Maps'].add_to(map)

df = pd.read_excel('/gdrive/MyDrive/andong_1.xlsx')
for i in range(len(df)):
  thereLoc = df['위도'][i], df['경도'][i]
  fo.CircleMarker(
      thereLoc,
      radius = 20,
      color = '#8000FF',
      fill_color = '#8000FF'
  ).add_to(map)

  fo.Marker(
      thereLoc,
      tooltip = df['장소명'][i],
      icon = fo.Icon(color = 'red', icon = 'heart' )
  ).add_to(map)
map
```

VWorld 배경지도API 활용 - 기본배경(.png형식)

```python
#라이브러리 가져오기
import pandas as pd
import folium as fo

lat, log = 36.543682613, 128.800027098
zoom = 13
map = fo.Map(
    location= [lat, log],
    zoom_start= zoom
)

# VWorld  배경지도 API 활용하기
VWMap = {
    'VWorldBase' : fo.TileLayer(
        tiles = 'http://api.vworld.kr/req/wmts/1.0.0/6D80D0CB-D1FE-3DD1-9C2F-6F8DD380CBCA/Satellite/{z}/{y}/{x}.jpeg',
        attr = 'VWorldBase',
        name = 'hello',
        overlay = True,
        control = True)
}
VWMap['VWorldBase'].add_to(map)

df = pd.read_excel('/gdrive/MyDrive/andong_1.xlsx')
for i in range(len(df)):
  thereLoc = df['위도'][i], df['경도'][i]
  fo.CircleMarker(
      thereLoc,
      radius = 20,
      color = '#8000FF',
      fill_color = '#8000FF'
  ).add_to(map)

  fo.Marker(
      thereLoc,
      tooltip = df['장소명'][i],
      icon = fo.Icon(color = 'red', icon = 'heart' )
  ).add_to(map)
# html 파일로 저장
map.save('/gdrive/MyDrive/andong_1.html')
```
